```python
top_countries = df1["country"].value_counts().head(10)

```

```python

plt.figure(figsize=(15,4))
sns.barplot(x=top_countries.index, y = top_countries.values)
plt.title('Countries with Most number of billionaires')
plt.xlabel('Country')
plt.ylabel('Number of Billionaires')
plt.show()
           
```

```python
gender = df1['gender'].value_counts().head()
```

```python
sns.barplot(x=gender.index, y = gender.values)
plt.title('Total number of Male/Female billionaires')
plt.xlabel('Male/Female')
plt.ylabel('Number of Billionaires')
plt.show()
```

```python
top_industries=df1['industries'].value_counts().head(10)
```

```python
plt.figure(figsize=(20,6))
sns.barplot(x=top_industries.index, y=top_industries.values)
plt.title('Top 10 industries with highest number of billionaires')
plt.xlabel('Industry')
plt.ylabel('Number of billionaires')
plt.show()
```
